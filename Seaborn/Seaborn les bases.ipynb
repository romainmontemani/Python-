# **1. Importation et installation du package**

```python
#Importation du package
import seaborn as sns
```

```python
#Liste des bases disponibles dans seaborn
sns.get_dataset_names()
```

```python
#Chargement de la base de données
iris = sns.load_dataset("iris")
iris.head()
```

```python
#Chargement de la base de donnée
titanic = sns.load_dataset("titanic")
titanic.head()
```

# **2.Graphiques simples avec Seaborn**

```python
#Nuage de points (scatterplot)
sns.scatterplot(x="sepal_length", y="petal_width", data=iris)
```

```python
#Ajouter des catégories
sns.scatterplot(x="sepal_length", y="petal_width", data=iris,hue="species")
```

```python
#Faire changer la taille des graphiques en fonction des valeurs d'une autre variable
sns.scatterplot(x="sepal_length", y="petal_width", data=iris,hue="species",size="petal_length")
```

```python
#Changer la palette de couleur
sns.scatterplot(x="sepal_length", y="petal_width", data=iris,hue="species",size="petal_length",palette="YlGnBu")
```

# **3.Quelques graphiques à toujours réaliser dans l'analyse de donnée**

**3.1 Analyser toutes les variables grâce au pairplot**

```python
#pairplot basique
sns.pairplot(iris)
```

```python
#Analyse du pairplot suivant une catégorie
sns.pairplot(iris, hue="species", kind="kde")
```

**3.2 Histogramme**

```python
#Tacer un histogramme avec la courbe de densité
sns.histplot(iris.petal_length)
```

**3.3 Barplot**

```python
#Barplot
sns.barplot(x="species", y="petal_length", data=iris, palette="YlGnBu")
```

**3.4 Boxplot**

```python
#Boxplot entre une qualtitative et une quantitative
sns.boxplot(x="species", y="petal_length", data=iris, palette="YlGnBu")
```
