**Exercice Numpy** :

Supposons que nous disposons d'un jeu de données contenant les scores obtenus par des étudiants lors d'un examen. Notre objectif sera d'utiliser nos compétences en Numpy pour répondre à plusieurs questions à leurs sujets.

**1. Importer Numpy :**

```python
import numpy as np
```

Considérons ensuite les notes suivantes :

```python
notes = [15,17,10,20,15.5,18,13,14,19,12,11,9,8]
```

**2. À partir de ces notes, créez un array que vous nommerez arr :**

```python
arr=np.array(notes)
arr
```

**3. Calculer la moyenne des notes en classe**

```python
moyenne=np.mean(arr)
moyenne
```

**4. Trouvez le minimum et le maximum de la classe**

```python
minimum=np.min(arr)
minimum
```

```python
maximum=np.max(arr)
maximum
```

**5. Trouver le nombre d'étudiants ayant obtenu une note supérieur à la moyenne :**

```python
n=np.sum(arr>moyenne)
n
```

**6. Modifiez les notes inférieures à 10 en 10**

```python
arr[arr<10]=10
arr
```
