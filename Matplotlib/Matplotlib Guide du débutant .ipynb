**1. Installation et importation du package**

```python
#Importation des packages
import matplotlib.pyplot as plt
import numpy as np
```

**2. Tracé de graphiques simple avec Matplotlib**

```python
#Génération de données
x=np.linspace(0, 10, 1000)
y=np.exp(x)
```

```python
#Tracer un graphique simple
plt.plot(x,y)
```

**3. Personnaliostion de graphique avec Matplotlib**

```python
#Changer la taille d'un graphique avec la largeur et la hauteur de la figure en pouce
plt.figure(figsize=(5,5))
plt.plot(x,y)
```

```python
#Titres et axes
plt.plot(x,y)
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
```

```python
#Changement de couleur
plt.plot(x,y,c="green")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
```

```python
#Changer le type de ligne
plt.plot(x,y,c="green",ls="--")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
```

```python
#Ajouter une légende
plt.plot(x,y,c="green",ls="--",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
```

```python
#Ajouter une grille sur le graphique
plt.plot(x,y,c="green",ls="--",label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid()
```

**4. Faire plusieurs graphiques**

**4.1 Faire des graphiques sur un même carré**

```python
#Créer un graphique vide
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="--",label="exp")
plt.plot(x,z,c="blue",ls="--",label="2*exp")
plt.plot(x,y3,c="green",ls="--",label="4*exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid()
```

```python
#Changer les limites des axes
z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="--",label="exp")
plt.plot(x,z,c="blue",ls="--",label="2*exp")
plt.plot(x,y3,c="green",ls="--",label="4*exp")
plt.xlim((6,10))
plt.ylim((10000,80000))
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("Valeur")
plt.legend()
plt.grid()
```

**4.2 Faire des subplots**

**4.2.1 Approche fonctionnelle**

```python
#Graphique avec subplot (+ ajuster l'écart avec subplots_adjust)
plt.subplot(2,2,1)
plt.plot(x,y,c="red",ls="--",label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="--",label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="--",label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="--",label="exp")
plt.subplots_adjust(wspace=0.5)
```

**4.2.2 Approche orientée objet**

```python
#subplots
fig,axes=plt.subplots(2,1)
axes[0].plot(x,y,c="red")
axes[1].plot(x,y,c="blue")
```

**5. Sauvegarder le graphique**

```python
#Sauvegarder son graphique dans un fichier (savefig)
plt.subplot(2,2,1)
plt.plot(x,y,c="red",ls="--",label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="--",label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="--",label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="--",label="exp")
plt.subplots_adjust(wspace=0.5)
plt.savefig("figure.png")
plt.savefig("figure.jpeg")
```
